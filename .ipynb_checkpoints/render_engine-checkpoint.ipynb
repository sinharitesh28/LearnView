```python
import json
import os

# --- Configuration ---
MANIFEST_PATH = 'layout_manifest.json'
OUTPUT_SVG_PATH = 'final_slide_output.svg'
CANVAS_WIDTH = 1920
CANVAS_HEIGHT = 1080

def style_dict_to_css(style_dict):
    """Converts a dictionary of styles into an inline CSS string."""
    return "; ".join([f"{k}: {v}" for k, v in style_dict.items()])

def estimate_bounding_box(element):
    """Calculates approximate bounding boxes (X, Y, Width, Height) for layout analysis."""
    geo = element.get('geometry_attrs', {})
    obj_type = element.get('object_type')
    style = element.get('style_attrs', {})
    
    bbox = {"x": 0, "y": 0, "w": 0, "h": 0}
    
    if obj_type in ['rect', 'image']:
        bbox = {"x": geo.get('x', 0), "y": geo.get('y', 0), "w": geo.get('width', 0), "h": geo.get('height', 0)}
        
    elif obj_type == 'text':
        # Heuristic estimation for text bounds
        font_size_str = style.get('font-size', '30px')
        font_size = int(''.join(filter(str.isdigit, font_size_str)))
        text_length = len(element.get('content', ''))
        
        # Approx width: ~0.55 of font size per character
        bbox['w'] = text_length * (font_size * 0.55)
        bbox['h'] = font_size
        bbox['y'] = geo.get('y', 0) - font_size  # SVG text Y is usually the baseline
        
        # Adjust X based on text-anchor
        bbox['x'] = geo.get('x', 0)
        if style.get('text-anchor') == 'middle':
            bbox['x'] -= (bbox['w'] / 2)
            
    elif obj_type == 'line':
        x1, y1 = geo.get('x1', 0), geo.get('y1', 0)
        x2, y2 = geo.get('x2', 0), geo.get('y2', 0)
        bbox = {"x": min(x1, x2), "y": min(y1, y2), "w": abs(x2 - x1), "h": max(abs(y2 - y1), 2)} # min 2px height for horizontal lines
        
    return bbox

def analyze_layout(manifest_data):
    """Generates a summary of spatial usage and detects layout anomalies."""
    elements = manifest_data.get('layout_manifest', [])
    summary = {
        "canvas": {"width": CANVAS_WIDTH, "height": CANVAS_HEIGHT},
        "total_elements": len(elements),
        "element_metrics": [],
        "warnings": []
    }
    
    for el in elements:
        obj_id = el.get('object_id', 'unknown')
        bbox = estimate_bounding_box(el)
        
        metrics = {
            "id": obj_id,
            "type": el.get('object_type'),
            "bounding_box": bbox
        }
        summary["element_metrics"].append(metrics)
        
        # Check if element bleeds off canvas (ignoring background elements)
        if el.get('z_order', 0) > 0 and obj_id != "header_banner":
            if bbox['x'] < 0 or bbox['y'] < 0 or (bbox['x'] + bbox['w'] > CANVAS_WIDTH) or (bbox['y'] + bbox['h'] > CANVAS_HEIGHT):
                summary["warnings"].append(f"Element '{obj_id}' is out of bounds.")
                
    return summary

def generate_svg(manifest_data, output_path):
    """Compiles the manifest data into an SVG file."""
    elements = manifest_data.get('layout_manifest', [])
    
    # Sort elements by Z-order to ensure correct rendering stacking
    elements = sorted(elements, key=lambda x: x.get('z_order', 0))
    
    svg_lines = [
        f'<svg xmlns="http://www.w3.org/2000/svg" viewBox="0 0 {CANVAS_WIDTH} {CANVAS_HEIGHT}" width="{CANVAS_WIDTH}" height="{CANVAS_HEIGHT}">',
        '  '
    ]
    
    for el in elements:
        obj_type = el.get('object_type')
        obj_id = el.get('object_id')
        geo = el.get('geometry_attrs', {})
        style = el.get('style_attrs', {})
        style_css = style_dict_to_css(style)
        
        tag = ""
        if obj_type == 'rect':
            tag = f'  <rect id="{obj_id}" x="{geo.get("x")}" y="{geo.get("y")}" width="{geo.get("width")}" height="{geo.get("height")}" style="{style_css}" />'
        elif obj_type == 'text':
            content = el.get('content', '')
            tag = f'  <text id="{obj_id}" x="{geo.get("x")}" y="{geo.get("y")}" style="{style_css}">{content}</text>'
        elif obj_type == 'image':
            href = el.get('href', '')
            # Update extension from jpeg to png to match the asset directory instructions if needed
            if "jpeg" in href:
                href = href.replace(".jpeg", ".png")
            tag = f'  <image id="{obj_id}" x="{geo.get("x")}" y="{geo.get("y")}" width="{geo.get("width")}" height="{geo.get("height")}" href="{href}" style="{style_css}" />'
        elif obj_type == 'line':
            tag = f'  <line id="{obj_id}" x1="{geo.get("x1")}" y1="{geo.get("y1")}" x2="{geo.get("x2")}" y2="{geo.get("y2")}" style="{style_css}" />'
        
        if tag:
            svg_lines.append(tag)
            
    svg_lines.append('</svg>')
    
    with open(output_path, 'w', encoding='utf-8') as f:
        f.write("\n".join(svg_lines))
        
    print(f"✅ Successfully compiled SVG to: {output_path}")

# --- Main Execution ---
if __name__ == "__main__":
    if not os.path.exists(MANIFEST_PATH):
        print(f"❌ Error: {MANIFEST_PATH} not found.")
    else:
        # Load Manifest
        with open(MANIFEST_PATH, 'r', encoding='utf-8') as f:
            manifest = json.load(f)
            
        # 1. Generate the visual SVG
        generate_svg(manifest, OUTPUT_SVG_PATH)
        
        # 2. Generate and print the analytical summary for the AI
        layout_summary = analyze_layout(manifest)
        print("\n--- AI Layout Analysis Summary ---")
        print(json.dumps(layout_summary, indent=2))
```
